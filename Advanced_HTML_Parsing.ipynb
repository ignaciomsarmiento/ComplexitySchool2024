## Advanced HTML Parsing


### BeautifulSoup and stylesheets

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen('https://www.pythonscraping.com/pages/warandpeace.html')
bs = BeautifulSoup(html, 'html.parser')

print(bs.prettify())
```

```python
nameList = bs.find_all('span',{'class': ['green','red']})

for name in nameList:
    print(name.get_text())
```

### Navigating Trees

```python
html_gift = urlopen('https://www.pythonscraping.com/pages/page3.html')

bs_gift = BeautifulSoup(html_gift, 'html.parser')

for child in bs_gift.find('table',{'id':'giftList'}).children:
    print(child)
```

### Extracting a Table from Wikipedia

```python
import pandas as pd

html_wiki = urlopen('https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)')

bs_wiki = BeautifulSoup(html_wiki,'html.parser')
```

```python
table = bs_wiki.find('table',{'class':'wikitable'})
```

```python
rows = []

for row in table.find_all('tr'):
    cells = row.find_all(['th', 'td'])
    cell_text = [cell.get_text(strip=True) for cell in cells]
    rows.append(cell_text)
```

```python
df = pd.DataFrame(rows)

df
```
